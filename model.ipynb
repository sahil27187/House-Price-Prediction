```python
#convert total_sft range  to float 
def convertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1])) / 2
    try:
        return float(x)
    except:
        return None
    
```

```python
data['total_sqft']=data['total_sqft'].apply(convertRange);
```

```python
#proce per square feet calculatoin

data['price_per_sqft'] = data['price']*100000/data['total_sqft']
```

```python
# we need to reduce the data set so that we can train so less than occurace of any location we will consider it as other 
data['location'] = data['location'].apply(lambda x:x.strip()) # remeove extra space 

location_count = data['location'].value_counts()
location_count
```

```python
location_count_less_10 = location_count[location_count<=10];
location_count_less_10
```

```python
# so less than 10 we will consider as other 
data['location'] = data['location'].apply(lambda x : 'other' if x in location_count_less_10 else x)

#check it chnaged or not 
data['location'].value_counts()
```

```python
data.describe()
```

```python
# now we have some outliers to remove like total_sft is 1.00000 is min 
# try total_sqft per bhk 
(data['total_sqft']/data['bhk']).describe()
```

```python
# it is not possible to get a flat of 1bhk for 0.25sqft
#so set the range
data = data[((data['total_sqft']/data['bhk'])>=300)]
data.describe()
```

```python
# to remove the outliers we will be making a function that hold group by data based on location and for each loacation gorup by based on the bhk .
# for each bhk we will create a dict having it mean,std,count, and keep the only data with validate the minimum of the cost of 2bhk mean for 3 bhk
#also there should be atleast 5 data points
# # else drop it, 
def bhk_outlier_remover(df):
    exclude_indices = np.array([])#store the data be to excluded
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < stats['mean']].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
data = bhk_outlier_remover(data)
```

```python
data.shape
# price per sqft was only feature which helped to remove the outliers.
```

```python
#exclued price from x and store in y 
X = data.drop(columns=['price']);
y = data['price']
```

```python
#apply three models linear regression , lasso,ridge and preporccseing based on one hot incoder, standard scallar 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
#apply train _test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape);
print(X_test.shape)
```

```python
#apply one hot incoder on location only becousr it was catgorical
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['location']), remainder='passthrough')
```

```python
#to slace all the things 
scaler = StandardScaler()
```

```python
#initialize linear regression 
lr = LinearRegression()
```

```python
#pipline provide the sequencelial exectuion of the coloumn_trrans then scaller , then linear regression 
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
#now predict the result
y_pred_lr = pipe.predict(X_train)
```

```python
# get the r2 score for y_pred_lr and y_test
r2_score(y_train,y_pred_lr) # this is using the linear regression 
```

```python
# now apply lasso 
lasso= Lasso(); 
```

```python
#arrange the sequ you want to run this pipeline
pipe = make_pipeline(column_trans,scaler,lasso)
```

```python
#fit the dataset 
pipe.fit(X_train,y_train)
```
